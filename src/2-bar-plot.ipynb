### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
# Import the required library
import matplotlib.pyplot as plt

try:
    # Data
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    colors = ['red', 'blue', 'green', 'orange']  # Different colors for each bar

    # Validate data length
    if len(categories) != len(values) or len(categories) != len(colors):
        raise ValueError("Mismatch in the length of categories, values, or colors.")

    # Create the bar plot
    plt.bar(categories, values, color=colors)

    # Add title and labels
    plt.title("Bar Plot of Categories", fontsize=14)
    plt.xlabel("Categories", fontsize=12)
    plt.ylabel("Values", fontsize=12)

    # Show the plot
    plt.show()

except Exception as e:
    print(f"An error occurred: {e}")
```
