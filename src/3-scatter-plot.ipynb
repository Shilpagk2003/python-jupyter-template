### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
# Import the required library
import matplotlib.pyplot as plt

try:
    # Data
    x = [1, 2, 3, 4, 5]
    y = [2, 4, 6, 8, 10]

    # Validate data length
    if len(x) != len(y):
        raise ValueError("Mismatch in the length of x and y data.")

    # Validate data types
    if not all(isinstance(i, (int, float)) for i in x + y):
        raise TypeError("All elements in x and y must be integers or floats.")

    # Create scatter plot
    plt.scatter(x, y, color='blue', label='Data Points')

    # Annotate each point
    for i, j in zip(x, y):
        plt.text(i, j, f'({i}, {j})', fontsize=10, ha='right', va='bottom', color='red')

    # Add title and labels
    plt.title("Scatter Plot with Annotations", fontsize=14)
    plt.xlabel("X Values", fontsize=12)
    plt.ylabel("Y Values", fontsize=12)

    # Add a legend
    plt.legend()

    # Display the plot
    plt.show()

except Exception as e:
    print(f"An error occurred: {e}")
```
