### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
# Import the required library
import matplotlib.pyplot as plt

try:
    # Data for the plots
    line_x = range(0, 11)
    line_y = [i**2 for i in line_x]

    bar_categories = ['A', 'B', 'C', 'D']
    bar_values = [5, 7, 3, 9]

    scatter_x = [1, 2, 3, 4, 5]
    scatter_y = [2, 4, 6, 8, 10]

    pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
    pie_sizes = [40, 25, 20, 15]

    # Validate data
    if len(bar_categories) != len(bar_values):
        raise ValueError("Mismatch between bar categories and values.")

    if len(scatter_x) != len(scatter_y):
        raise ValueError("Mismatch between scatter x and y data.")

    if len(pie_labels) != len(pie_sizes):
        raise ValueError("Mismatch between pie labels and sizes.")

    if sum(pie_sizes) <= 0:
        raise ValueError("Sum of pie chart sizes must be greater than zero.")

    # Create the 2x2 subplot layout
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))

    # First subplot: Line chart
    axes[0, 0].plot(line_x, line_y, color='blue', marker='o', label="y = x^2")
    axes[0, 0].set_title("Line Chart")
    axes[0, 0].set_xlabel("X Values")
    axes[0, 0].set_ylabel("Y Values")
    axes[0, 0].legend()
    axes[0, 0].grid(True)

    # Second subplot: Bar chart
    axes[0, 1].bar(bar_categories, bar_values, color=['red', 'green', 'blue', 'orange'])
    axes[0, 1].set_title("Bar Chart")
    axes[0, 1].set_xlabel("Categories")
    axes[0, 1].set_ylabel("Values")

    # Third subplot: Scatter plot
    axes[1, 0].scatter(scatter_x, scatter_y, color='purple', label="Data Points")
    axes[1, 0].set_title("Scatter Plot")
    axes[1, 0].set_xlabel("X Values")
    axes[1, 0].set_ylabel("Y Values")
    axes[1, 0].legend()

    # Fourth subplot: Pie chart
    axes[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140,
                   colors=['gold', 'skyblue', 'lightgreen', 'salmon'])
    axes[1, 1].set_title("Pie Chart")

    # Adjust layout for better spacing
    plt.tight_layout()

    # Display the plots
    plt.show()

except Exception as e:
    print(f"An error occurred: {e}")
```
