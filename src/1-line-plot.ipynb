### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
# Import the required library
import matplotlib.pyplot as plt

# Define the data
x = range(0, 11)  # x values ranging from 0 to 10
y = [i**2 for i in x]  # y values as the square of x

# Create the line plot
plt.plot(x, y, marker='o', color='blue', label='y = x^2')

# Add customization
plt.title("Line Plot: y = x²", fontsize=14)  # Title of the plot
plt.xlabel("X Values", fontsize=12)         # X-axis label
plt.ylabel("Y Values (x²)", fontsize=12)    # Y-axis label
plt.grid(True, linestyle='--', alpha=0.7)   # Add a grid with dashed style

# Display the plot
plt.legend()  # Add a legend
plt.show()
```
